```python
import pandas as pd

# Load your dataset
df = pd.read_csv(FILE)

# Compute average weather per Resort + SkiWeek across all past years
avg_weather = df.groupby(['Resort', 'SkiWeek']).agg({
    'Maximum temperature (Degree C)': 'mean',
    'Minimum temperature (Degree C)': 'mean',
    'Rainfall amount (millimetres)': 'mean'
}).reset_index()

# Add Year = 2026 for prediction
avg_weather['Year'] = 2026

# Reorder columns to match original dataset format
pred_2026 = avg_weather[['Year', 'SkiWeek', 'Resort',
                         'Maximum temperature (Degree C)',
                         'Minimum temperature (Degree C)',
                         'Rainfall amount (millimetres)']]

# Save to CSV
pred_2026.to_csv("predicted_weather_2026.csv", index=False)
```
